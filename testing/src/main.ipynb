#### Perform imports and declaration of globals

```python
import scapy
from scapy.all import *
```

Read the pcap and save to variable 'pkts'

```python
pkts = rdpcap('packet.pcap')
```

#### Timeshift the whole pcap one month back in time
Attempt to change the time of the pcap but keeping the delta between packets

```python
oneMonth = 2630000
newPackets=[]

# For all packets, subtract one month (30 days) from the time (unix in seconds), add to newPackets
for packet in pkts:
  packet.time = packet.time - oneMonth
  packetModified=packet
  newPackets.append(packetModified)

# Write the new manipulated packets to 'dump.pcap'
wrpcap("./dump.pcap",newPackets)
```

#### Separate the pcap's sessions/flows from each other

```python
flows = pkts.sessions()

# Extract key and packets for a session
key = next(iter(flows))
packets = flows[key]

# For all packets in session
#for p in packets:
 # print(p.time)
  
for flow, packets in flows.items():
  print(flow)
  for packet in packets:
    print(packet)
```
